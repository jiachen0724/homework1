```python
import numpy as np
import matplotlib.pyplot as plt

# 第一組參數：a = -3, b = 4, c = 7
a1 = -3
b1 = 4
c1 = 7

# 第二組參數：a = 5, b = -4, c = 9
a2 = 5
b2 = -4
c2 = 9

# 設定x的範圍
x = np.linspace(-10, 10, 400)

# 計算y值
y1 = a1 * x**2 + b1 * x + c1
y2 = a2 * x**2 + b2 * x + c2

# 繪製圖形
plt.plot(x, y1, label="Y = -3X^2 + 4X + 7")
plt.plot(x, y2, label="Y = 5X^2 - 4X + 9")

# 找出根的位置
roots1 = np.roots([a1, b1, c1])
roots2 = np.roots([a2, b2, c2])

# 找出最小值的位置
vertex1 = (-b1) / (2 * a1)
vertex2 = (-b2) / (2 * a2)
min_value1 = a1 * vertex1**2 + b1 * vertex1 + c1
min_value2 = a2 * vertex2**2 + b2 * vertex2 + c2

# 標注根和最小值
plt.scatter(roots1, [0, 0], color='red', label='Roots')
plt.scatter(roots2, [0, 0], color='red')
plt.scatter(vertex1, min_value1, color='green', label='Minimum')
plt.scatter(vertex2, min_value2, color='green')

# 添加圖例
plt.legend()

# 顯示圖形
plt.show()
```
